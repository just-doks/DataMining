```python
import numpy as np
s = np.array([0.0]*11);
s[0] = 1.2
for i in range(10):
  if (s[i] > 0.1):
    s[i+1] = (s[0]/(i+1))
  if (s[i] <= 0.1):
    s[i+1] = 0.1
print(s)
```

```python
import numpy as np
import random
import math
#функция f(x,y,z)
def f(x):
  return ((3*x[0]**2 +5*x[1]**2)/(math.sin(x[2])+2))
#частные производные функции
def grad(x):
  return np.array([6*x[0]/(math.sin(x[2])+2), 10*x[1]/(math.sin(x[2])+2),
                   -(((3*x[0]**2+5*x[1]**2)*math.cos(x[2]))/(math.sin(x[2]+2)**2))])

# Стартовая точка
x0 = random.randint(-5,5)
x1 = random.randint(-5,5)
x2 = random.randint(-5,5)
x = np.array([x0, x1, x2])
print("начальная точка: ")
print(x0)
print(x1)
print(x2)

max_iter = 10
eps = 1e-4

f_old = f(x)
x = x - grad(x) * s[0]
f_new = f(x)
i = 0

while abs(f_old - f_new) > eps and i < max_iter:
  i = i + 1
  f_old = f_new
  print(f(x))
  x = x - grad(x) * s[i]
  f_new = f(x)

print(f(x))
```
