```python
import pandas as pd
data = pd.read_csv("./lr2-dataset.csv")
data
```

```python
# Среднее значение
mymean = data.mean(axis=0, numeric_only=False)
# Стандартное отклонение
mystd = data.std(axis=0, numeric_only=False)
data -= mymean
data /= mystd
data
```

```python
X = data.iloc[:,0]
Y = data.iloc[:,1]
Z = data.iloc[:,2]
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y)
plt.show()
```

```python
ax = plt.axes(projection='3d')
ax.scatter(X, Y, Z)
plt.show()
```

```python
n, bin, patches = plt.hist(Y, bins=20)
plt.show()
```

```python
n, bin, patches = plt.hist(Z, bins=20)
plt.show()
```
