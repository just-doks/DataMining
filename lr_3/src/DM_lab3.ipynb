```python
import pandas as pd
data = pd.read_csv("./biopsy.csv")
data
```

```python
data.info()
```

```python
data.drop_duplicates(keep = 'first', inplace=True)
data.dropna(axis = 0, inplace = True)
data.info()
```

```python
data = data.drop(columns = ['rownames'])
data
```

```python
import matplotlib.pyplot as plt
plt.boxplot(data['V1'])
```

```python
data = data.drop(columns=['ID'])
data
```

```python
x = data.iloc[:,0:9]
x
```

```python
y = data.iloc[:,9:10]
y
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
subdata_encoded = y.apply(le.fit_transform)
subdata_encoded
```

```python
subdata_encoded = data.iloc[:,0]
subdata_encoded
```

```python
import matplotlib.pyplot as plt
plt.boxplot(x,notch=True)
```

```python
import seaborn as sns
sns.heatmap(x.corr(), annot = True)
```

```python
x.corr()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
X_train, X_test, y_train, y_test = train_test_split(x, subdata_encoded, test_size=0.33, random_state=42)
model = Sequential()
model.add(Dense(32, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(128, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(1))
print(model.summary())
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split = 0.2, verbose=2)
```

```python
pred = model.predict(X_test)
```

```python
print(pred)
```
