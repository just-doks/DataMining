```python
import matplotlib.pyplot as plt

Data = [(1, 1), (-2, -2), (5, 7), (-3, -5), (0, 4)]
Clusters0 = [(0, 2), (-1, -1)]
Clusters1 = [(1.11, 2.74), (-1.04, -1.13)]
Clusters2 = [(1.34, 3.25), (-0.61, -1.59)]

for (x,y) in Data:
    plt.scatter(x, y, c='r')
    plt.annotate('Данные', xy=[x, y], xycoords='data', xytext=[6, -3], 
                 fontsize=8, textcoords='offset points')
for (x,y) in Clusters0:
    plt.scatter(x, y, c='g')
    plt.annotate('Кластеры(0)', xy=[x, y], xycoords='data', xytext=[-56, 3], 
                 fontsize=8, textcoords='offset points')
for (x,y) in Clusters1:
    plt.scatter(x, y, c='b')
    plt.annotate('Кластеры(1)', xy=[x, y], xycoords='data', xytext=[-58, -3], 
                 fontsize=8, textcoords='offset points')
for (x,y) in Clusters2:
    plt.scatter(x, y, c='y')
    plt.annotate('Кластеры(2)', xy=[x, y], xycoords='data', xytext=[6, -3], 
                 fontsize=8, textcoords='offset points')

plt.xticks(ticks=range(-6, 9))
plt.yticks(range(-6, 9))
plt.tick_params(axis='both', color='#a9a9a9', labelcolor='#a9a9a9')

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.spines['bottom'].set_color('#a9a9a9')
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
ax.spines['left'].set_color('#a9a9a9')
plt.show()
```
